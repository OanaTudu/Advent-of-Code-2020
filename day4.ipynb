```python
import pandas as pd
import math
import re
file1 = open("input4.txt","r+")  
text = file1.readlines() 
file1.close() 
```

```python
text
passports = []
p = ''
for el in text:
    if el != '\n':
        p += el
    else:
        p = p[:-1]
        p = p.replace('\n',' ')
        passports.append(p)
        p = ''
passports 
```

```python
def valid_pass(passport):
    """function to check whether a passport is valid or not"""
    valid = 0
    parts = passport.split(' ')
    #print(parts)
    if len(parts) == 8:
        #print('valid')
        valid = 1
    elif len(parts) < 7:
        #print('not valid')
        valid = 0
    else:
        keys = []
        for kvp in parts:
            #print(kvp.split(':'))
            keys.append(kvp.split(':')[0])
        #print(keys)
        if 'cid' in keys:
            #print('not valid')
            valid = 0
        else:
            #print('valid')
            valid = 1
    return valid
```

```python
nr_valid = 0
for pas in passports:
    nr_valid += valid_pass(pas)
nr_valid
```
