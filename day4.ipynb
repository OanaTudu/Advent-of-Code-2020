```python
def valid_pass(passport):
    """function to check whether a passport is valid or not"""
    valid = 0
    parts = passport.split(' ')
    #print(parts)
    if len(parts) == 8:
        #print('valid')
        valid = 1
    elif len(parts) < 7:
        #print('not valid')
        valid = 0
    else:
        keys = []
        for kvp in parts:
            #print(kvp.split(':'))
            keys.append(kvp.split(':')[0])
        #print(keys)
        if 'cid' in keys:
            #print('not valid')
            valid = 0
        else:
            #print('valid')
            valid = 1
    return valid
```

```python
nr_valid = 0
for pas in passports:
    nr_valid += valid_pass(pas)
nr_valid
```

```python
def pass_valid(passport_dict):
    """function to validate passport fields:
    Validation rules:
    byr (Birth Year) - four digits; at least 1920 and at most 2002.
    iyr (Issue Year) - four digits; at least 2010 and at most 2020.
    eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
    hgt (Height) - a number followed by either cm or in:
        If cm, the number must be at least 150 and at most 193.
        If in, the number must be at least 59 and at most 76.
    hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
    ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
    pid (Passport ID) - a nine-digit number, including leading zeroes.
    cid (Country ID) - ignored, missing or not."""
    valid = 1
    #check birth year:
    if 'byr' in passport_dict:
        #check rules:
        if len(passport_dict['byr']) != 4: 
            v = 0
            print('byr does not have 4 digits')
        elif int(passport_dict['byr']) < 1920 or int(passport_dict['byr']) > 2002: 
            v = 0    
            print('byr is outside the range')
        else:
            v = 1
    else:
        print('byr not in passport')
        return 0
    valid *= v
    #check issue year:
    if 'iyr' in passport_dict:
        #check rules:
        if len(passport_dict['iyr']) != 4: 
            print('iyr does not have 4 digits')
            v = 0
        elif int(passport_dict['iyr']) < 2010 or int(passport_dict['iyr']) > 2020: 
            print('iyr is outside range')
            v = 0  
        else:
            v = 1
    else:
        print('iyr not in passport')
        return 0
    valid *= v
    #check expiration year:
    if 'eyr' in passport_dict:
        #check rules:
        if len(passport_dict['eyr']) != 4: 
            print('eyr not 4 digits')
            v = 0
        elif int(passport_dict['eyr']) < 2020 or int(passport_dict['eyr']) > 2030: 
            print('eyr not in range')
            v = 0       
        else:
            v = 1
    else:
        print('eyr not in passport')
        return 0
    #check passport id:
    if 'pid' in passport_dict:
        #check rules:
        if len(passport_dict['pid']) != 9:
            print('pid does not have 9 characters length')
            v = 0
        elif passport_dict['pid'].isdigit():
            v = 1
        else:
            print('pid not digits')
            v = 0
    else:
        print('pid not in passport')
        return 0
    #check height:
    if 'hgt' in passport_dict:
        #check rules:
        h = int(re.findall(r'\d+',passport_dict['hgt'])[0])
        u = "".join(re.split("[^a-zA-Z]*", passport_dict['hgt'])) 
        if u == 'cm':
            if h < 150 or h > 193: 
                print('invalid height in cm')
                v = 0
            else:
                v = 1
        elif u == 'in':
            if h < 59 or h > 76: 
                print('invalid height in inches')
                v = 0
            else:
                v = 1
        else: 
            print('height in some unknown unit')
            v = 0
    else:
        print('hgt not in passport')
        return 0
    #check hair color:
    if 'hcl' in passport_dict:
        if passport_dict['hcl'][0] == '#' and len(passport_dict['hcl']) == 7:
            g = "".join(re.split("[^a-f0-9]*",passport_dict['hcl'][1:]))
            if g == dict_test['hcl'][1:]: 
                v = 1
            else:
                print('non-matching hcl')
        else:
            print('hcl wrong length or missing initial #')
            v = 0
    else:
        print('hcl not in passport')
        return 0
    #check eye color:
    if 'ecl' in passport_dict:
        #check if it's in valid list:
        #amb blu brn gry grn hzl oth
        if passport_dict['ecl'] in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']: 
            v = 1
        else:
            print('not a valid eye color string')
            v = 0
    else:
        print('ecl not in passport')
        return 0
    
    return v
        
```

```python
nr_valid = 0
for pas in passports:
    #turn passport to dictionary:
    fields = pas.split(' ')
    pas_dict = {}
    for kvp in fields:
        pas_dict[kvp.split(':')[0]] = kvp.split(':')[1]
    nr_valid += pass_valid(pas_dict)
nr_valid
```
